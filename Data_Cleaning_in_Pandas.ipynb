# Data Cleaning

## Importing files and Data

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("/content/drive/MyDrive/Data Science Practice/World population/Customer Call List.xlsx")
df
```

## Droping Duplicates and Columns

```python
df.drop_duplicates(inplace = True)
df
```

```python
df = df.drop(columns = "Not_Useful_Column")
df
```

## Strip Function

```python
# Strip Function is used to strip unwanted things from the string by default it only removes blank spaces however you can also add specifics to be removed
# from the string

df["Last_Name"].str.strip()
```

```python
# df["Last_Name"] = df["Last_Name"].str.lstrip("...")
# df["Last_Name"] = df["Last_Name"].str.lstrip("/")
# df["Last_Name"] = df["Last_Name"].str.rstrip("_")

#This is a more easier way to use strip for multiple symbols at the same time.

df["Last_Name"] = df["Last_Name"].str.strip("123./_")
df
```

## Cleaning and Standardizing Phone Numbers

```python
# ASTYPE() param is used to convert the column data type which in this case we are converting it to string

df["Phone_Number"] = df['Phone_Number'].astype(str)

# Here using the REPLACE function we are replacing small a-z caps A-Z and num from 0-9 and replacing them with blank i.e. with nothing
# This will remove all the unwanted contents from the data
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','')
df
```

```python
# The following method can also be used to convert the column to string
# df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))

# Using this lambda function we would add  and concatenate '-' at the required places

df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df
```

```python
# Then we replace the nan-- and Na-- with blanks

df["Phone_Number"] = df["Phone_Number"].str.replace('nan--','')
df["Phone_Number"] = df["Phone_Number"].str.replace('Na--','')
df
```

## Splitting Address Data

```python
# Using the split function we split the address column into diff columns where the seperator ',' is the 1st arg, no_of_splits is 2nd arg and expand = True
# is 3rd arg which shows the view of the df
#df = df.drop(columns = ['State_Name', 'Pin_Code'])
df[["Street_Name","State","ZIP_Code"]] = df["Address"].str.split(',', 2, expand = True)
df
```

## Standardizing Columns using Replace Function

```python
# Create a replacement dictionary so that you do not replace 'Yes' with 'Yeses' while replacing 'Y' with 'Yes' similar for N and No

replacement_dict_y = {'Y': 'Yes','N':'No'}
#replacement_dict_n = {'N': 'No'}
```

```python
# Use the REPLACE function and pass the replacement_dict created as arg

df["Paying Customer"] = df["Paying Customer"].replace(replacement_dict_y)
df["Paying Customer"] = df["Paying Customer"].replace(replacement_dict_y)

df["Do_Not_Contact"] = df["Do_Not_Contact"].replace(replacement_dict_y)
df["Do_Not_Contact"] = df["Do_Not_Contact"].replace(replacement_dict_y)
# You can do the same replacement using Numpy as well but that makes things a bit complex but below is an example for the same.
#df['Paying Customer'] = np.where(df['Paying Customer'] == 'Y', 'Yes', df['Paying Customer'])

df
```

## Fill Null Values

```python
df = df.replace('N/a','')
df = df.fillna('')
df
```

## Filtering down Rows of Data

```python
# Replaced the blank values with no
df["Do_Not_Contact"] = df["Do_Not_Contact"].replace('','No')
df
```

```python
# Filters and droped Data where Do_Not_Contact was Yes as we do not need those values in table
# Also we dropped the columns where there was no contact number mentioned

for x in df.index:
  if df.loc[x, "Do_Not_Contact"] == 'Yes' or df.loc[x, "Phone_Number"] == '' :
    df.drop(x, inplace = True)


# We could have also dropped the Rows from Phone_Number Column using DROPNA() below is example for the same
# df.dropna(subset = "Phone_Number", inplace = True)
```

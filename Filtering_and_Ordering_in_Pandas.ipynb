```python
df[df['Rank'] < 10]
```

```python
specific_countries = ['Bangladesh', 'Brazil']
df[df['Country'].isin(specific_countries)]
```

```python
df[df['Country'].str.contains('United')]
```

```python
df2 = df.set_index('Country')
df2
```

```python
#Filters based on items parameter where you have to be specific about the column name and also mention the axis as required.
#For this example as we were using the axis = 1, which is also the default axis we can mention multiple column names

df2.filter( items = ['Rank', 'CCA3'],axis = 1)
```

```python
#Filters based on ITEMS parameter where you have to be specific about the column name and also mention the axis as required.

df2.filter(items = ['Zambia'], axis = 0)
```

```python
#Filters based on LIKE parameters where you can mention a substring for the entire string and if it matches you will get results, the axis rule follows for any and all parameters

df2.filter(like = 'United', axis = 0)
```

```python
df[df['Rank'] < 10 ].sort_values(by = 'Rank', ascending = False)
```

```python
df2[df2['Rank'] <10].sort_values(by = ['Continent', 'Country'], ascending = [True, False])
```

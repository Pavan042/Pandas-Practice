# Group by and Aggregation

```python
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/Data Science Practice/World population/Flavors.csv")
df
```

```python
# Creating Group by object

group_by_bf = df.groupby('Base Flavor')

# MEAN function on Group by

group_by_bf.mean()
```

```python
#MAX function on Group by

group_by_bf.max()
```

```python
# MIN function on Group by

group_by_bf.min()
```

```python
# Running it directly without saving it in a varaiable

df.groupby('Base Flavor').mean()
```

```python
# COUNT function on Groupby

group_by_bf.count()
```

```python
# Using SUM function with NUMERIC_ONLY param as True

df.groupby('Base Flavor').sum(numeric_only = True)
```

```python
# # Using SUM function with NUMERIC_ONLY param as False

df.groupby('Base Flavor').sum(numeric_only = False)
```

```python
#Using AGG Funtion to Group by on multiple function at the same time such as sum, min, max, count, etc

df.groupby('Base Flavor').agg({'Flavor Rating': ['mean', 'max', 'count', 'sum']})
```

```python
# You can do the same for multiple columns at the same time with different function param

df.groupby('Base Flavor').agg({'Flavor Rating': ['mean', 'max', 'count', 'sum'], 'Texture Rating': ['count','sum', 'mean']})
```

```python
# Using GROUP BY on multiple columns

df.groupby(['Base Flavor','Liked']).mean()
```

```python
# Using AGG function on multiple Group by columns

df.groupby(['Base Flavor','Liked']).agg({'Flavor Rating': ['mean', 'max', 'count', 'sum']})
```

```python
# Using DESCRIBE funtion on Group by

df.groupby('Base Flavor').describe()
```

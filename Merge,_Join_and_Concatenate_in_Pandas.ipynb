

## Merge



```python
# Default merge operation that performs inner merge/join


df1.merge(df2)
```

```python
# Using the HOW param we can  specify the type of merge/join

df1.merge(df2, how = 'inner')
```

```python
# Here we got the same O/P but we provided on based on which column data we need the merge to happen

df1.merge(df2, how = 'inner', on = ['FellowshipID','FirstName'])
```

```python
# Compared to the above example here we can see that there are two FirstName columns as FirstName_x, Firstname_yy this is based on the two different column
# that we perfomed merge on, as we didn't mention the FirstName as the param to do the merge on it took the First name form the Left table as _X and
# Right table as _y, now these are called suffixes which are added to the column name, we can change them using the SUFFIX param.

df1.merge(df2, how = 'inner', on = 'FellowshipID')
```

```python
# This outer merge will O/P all the values in both the table
# And the missing values will be stated as NaN
# However is there is overlapping data it won't be duplicated

df1.merge(df2, how = 'outer')
```

```python
# This performs a Left outer merge/join, where all the data from the left table is pulled and only the common data from the right is pulled

df1.merge(df2, how = 'left')
```

```python
# This performs a Right outer merge/join, where all the data from the right table is pulled and only the common data from the left is pulled

df1.merge(df2, how = 'right')
```

## Join


```python
# Using the join function we get similar results to merge, however here you have to mention all the paramerters for the statement to run.
# Params such as ON, HOW, LSUFFIX, RSUFFIX
# However Using joins is better when we perform join operations based on Index values

df1.join(df2, on = 'FellowshipID', how = 'outer', lsuffix = '_left', rsuffix = '_right')
```

```python
# Join based on INDEX which look more clean and easy to work with
# Here we use SET_INDEX param to set a column name as index from both the table

df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix = '_left', rsuffix = '_right')
df4
```

```python
# Here also we can use HOW param that specifies the type of join to be perfomed on the database.
# Below is the example of OUTER the same can be done for all the joins

df5 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix = '_left', rsuffix = '_right', how = 'outer')
df5
```

## Concatenate

```python
# Concatenate function stacks tables on top of one another which also means it does not join or merge based on any details.
# Hence you can get repeatative values

pd.concat([df1,df2])
```

```python
# Here we do have a JOIN param INNER which will only concatenate the columns with same name, where as OUTER will have the same results as default concat

pd.concat([df1,df2], join = 'inner')
```

```python
# Append works the same as concat in ways but as it is deprecated will use Concat instead if and as required

df1.append(df2)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df =  pd.read_csv("/content/drive/MyDrive/Data Science Practice/World population/Ice Cream Ratings.csv")
df = df.set_index('Date')
df.head()
```

```python
print(plt.style.available)
#plt.style.use('dark_background')
```

## Line Chart

```python
# By Default it will produce a line chart in order to produce diff charts use the param KIND

df.plot()
```

```python
# By adding SUBPLOT param we can create subplots of the same i.e. individual plots of the entire plot

df.plot(kind = 'line', subplots = True)
plt.show()
```

```python
# Adding title and labels using TITLE, XLABEL, YLABEL param

df.plot(kind = 'line', title = 'Ice Cream Ratings', xlabel = 'Dails Ratings', ylabel = 'Scores')
plt.show()
```

## Bar Chart

```python
df.plot(kind = 'barh')
plt.show()
```

```python
# Use STACKED param to stack it on top of each other

df.plot(kind = 'bar', stacked = True)
```

```python
# You can also create horizontal bar charts using any of the following two ways

# df.plot(kind = 'barh', stacked = True)

df.plot.barh(stacked = True)
```

```python
# Use SUBPLOTS to create subplots

df.plot(kind = 'barh', subplots = True)
plt.show()
```

```python
# Use Specific Coulmn name to create individual charts

df['Flavor Rating'].plot(kind = 'barh')
```

## Scatter Plot

```python
# Scatter plot requires two params by default that is X and Y.
# Param S is used to change the size of the dots.
# Param C is used to change the color of the dots

df.plot.scatter(x= 'Texture Rating', y= 'Overall Rating', s = 500, c = 'Red')
```

## Histogram

```python
df.plot.hist(bins = 20)
```

## Box Plot

```python
df.plot.box()
```

```python
df.boxplot()
```

## Area Plot

```python
df.plot.area(figsize = (10,5))
```

## Pie Chart

```python
df.plot.pie(y='Flavor Rating', figsize = (10,6))
```
